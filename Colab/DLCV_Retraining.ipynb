```python
import os
import shutil
from pathlib import Path

# 1. Setup paths
default_ckpt_path = Path("outputs/checkpoints/best_efficientnet_b0_sam.pt")
output_heatmap = Path("outputs/results/explainability/latest_pipeline_result.jpg")

# 2. Tournament Models
tournament_models = [
    ("Adam_EffNet", "outputs/checkpoints/best_effnet_adam_augmented.pt"),
    ("SAM_EffNet", "outputs/checkpoints/best_effnet_sam_augmented.pt"),
    ("ResNet50_Adam", "outputs/checkpoints/best_resnet50_adam_augmented.pt")
]

print(f"--- Starting Grad-CAM Tournament on /content/n2.jpg ---\n")

for name, augmented_file in tournament_models:
    if os.path.exists(augmented_file):
        print(f"\n{'='*50}\nTESTING MODEL: {name}\n{'='*50}")

        # Step A: Backup and Swap
        backup_created = False
        if default_ckpt_path.exists():
            os.rename(default_ckpt_path, "outputs/checkpoints/temp_backup.pt")
            backup_created = True

        try:
            shutil.copy(augmented_file, default_ckpt_path)

            # Step B: Run Pipeline
            !python pipeline.py /content/n2.jpg

            # Step C: Save unique result
            final_save_name = f"tournament_result_{name}.jpg"
            if output_heatmap.exists():
                shutil.copy(output_heatmap, final_save_name)
                print(f"\n[SUCCESS] Heatmap saved as: {final_save_name}")

        except Exception as e:
            print(f"\n[!] Error testing {name}: {e}")

        finally:
            # Step D: Cleanup and Restore original state
            if default_ckpt_path.exists():
                os.remove(default_ckpt_path)
            if backup_created:
                os.rename("outputs/checkpoints/temp_backup.pt", default_ckpt_path)
    else:
        print(f"[!] Skipping {name}: {augmented_file} not found.")

print("\n--- Tournament Finished! ---")
```
